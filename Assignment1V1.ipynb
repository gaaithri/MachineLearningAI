### Web Scraping the Countries where English is Official language  using  Beautiful Soup

```python
##import the library we use to open URLs
import urllib.request
```

```python
url = "https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language"
```

```python
# open the url using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(url)
```

```python
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup

```

```python
# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
# returns without title tag
soup.title.string
```

```python
doright_table=soup.findAll('table', class_ = 'wikitable sortable')
doright_table
```

```python
#Looping thru the list of different tables to extract countries

ccol = []
for i in doright_table:
    test_table = i
    for row in test_table.findAll('tr'):
        cells = row.findAll('td')
        if len(cells)==6:
            ccol.append(cells[2].find(text=True))
#             dcol.append(cells[3].find(text=True))
#             fcol.append(cells[5].find(text=True))
  
```

```python
import pandas as pd 
dftab = pd.DataFrame (ccol,columns = ['country_code'])

dftab = pd.concat([dftab[col].astype(str).str.lower() for col in dftab.columns],axis =1)
dftab = dftab.replace('\n','', regex=True) 


# dftab['Region']=dcol
# dftab['Primary_Eng_lang']=fcol
dftab.head()
```
